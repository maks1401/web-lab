```python
import os
import yt_dlp

def download_video_vk(urls):
    for url in urls:
        # Создаем папку по имени URL
        folder_name = url.split('//')[-1].split('/')[0]  # Используем домен для имени папки
        folder_path = os.path.join(os.getcwd(), folder_name)

        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
        
        # Параметры для yt-dlp
        ydl_opts = {
            'outtmpl': os.path.join(folder_path, '%(title)s.%(ext)s'),  # Путь к сохранению видео
            'writeinfojson': True,  # Скачивание метаданных в формате JSON
            'writethumbnail': True,  # Скачивание миниатюры
            'format': 'best',  # Загружаем видео в наилучшем качестве
        }
        
        # Используем yt-dlp для скачивания видео и метаданных
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            try:
                # Скачиваем видео
                ydl.download([url])
                
                # Получаем метаданные
                info_dict = ydl.extract_info(url, download=False)
                
                # Сохраняем метаданные в текстовый файл
                metadata_file_path = os.path.join(folder_path, f"{info_dict['title']}_metadata.txt")
                with open(metadata_file_path, 'w', encoding='utf-8') as f:
                    f.write(str(info_dict))  # Сохраняем информацию в текстовом формате
                
                print(f"Видеоконтент и метаданные для {url} успешно загружены в папку {folder_name}")

            except Exception as e:
                print(f"Не удалось скачать видео с {url}: {e}")

# Пример списка URL ВКонтакте
urls = [
    "https://vk.com/video-210653751_456244280", 
    "https://vk.com/video-160360338_456240245" 
]

# Вызов функции для скачивания видео
download_video_vk(urls)


```
